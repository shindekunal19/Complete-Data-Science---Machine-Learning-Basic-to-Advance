# Decide Dependant and Independant Elements in Data Set

```python
x = df[['bmi']]
y = df['charges'] # dependent variable
```

```python
x
```

```python
y
```

#Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(x,y,test_size=0.3)
```

```python
X_train.info()
```

```python
X_test.info()
```

# Select Model ML

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

# Trainig

```python
lm.fit(X_train, y_train )  # Training
```

#Testing

```python
predictions = lm.predict(X_test)  # Testing
```

# Model Evaluation

# Error Calcuation

Slide1.JPG

```python
import numpy as np   # Evaluation
import sklearn.metrics as metrics
print('MAE: {}'.format(metrics.mean_absolute_error(y_test, predictions)))
print('MSE: {}'.format(metrics.mean_squared_error(y_test, predictions)))
print('RMSE: {}'.format(np.sqrt(metrics.mean_squared_error(y_test, predictions))))
```

# Comapre Train, Test and Predicted Data

```python
import numpy as np
import plotly.graph_objects as go

# Assuming you have defined and trained your linear regression model (lm) and prepared your input data (X_train, y_train, X_test, y_test)

x_train = X_train.squeeze()
x_test = X_test.squeeze()

# Create an array of x values that spans the range of your training and test data
x = np.linspace(min(x_train.min(), x_test.min()), max(x_train.max(), x_test.max()), 100)

# Predict the y values for the generated x values
y = lm.predict(x.reshape(-1, 1))

fig = go.Figure()
fig.add_trace(go.Scatter(x=x_train, y=y_train, mode='markers', name='Train'))  # Training data
fig.add_trace(go.Scatter(x=x_test, y=y_test, mode='markers', name='Test'))    # Test data
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Prediction'))         # Predicted values

fig.update_layout(title='Linear Regression Model Visualization',
                  xaxis_title='X Axis Label',
                  yaxis_title='Y Axis Label')

fig.show()
```
