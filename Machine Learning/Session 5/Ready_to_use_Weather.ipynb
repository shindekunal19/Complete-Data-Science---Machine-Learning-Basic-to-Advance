```python
import pandas as pd
```

```python
import numpy as np
```

```python
air = pd.read_csv('air_store_info_with_nearest_active_station.csv')
```

```python
air_info = pd.read_csv('../air_store_info.csv')
```

```python
dict1 = air.set_index('air_store_id')['station_id'].to_dict()
```

```python
calendar_date = []
low_temperature =[]
avg_temperature = []
high_temperature = []
precipitation = []
air_id = []
for i in air_info.air_store_id.values:
    print i,dict1[i] + '.csv'
    k = pd.read_csv('1-1-16_5-31-17_Weather/'+dict1[i] + '.csv')
    for enu,j in enumerate(k.calendar_date.values):
        calendar_date.append(j)
        low_temperature.append(k.low_temperature.values[enu])
        avg_temperature.append(k.avg_temperature.values[enu])
        high_temperature.append(k.high_temperature.values[enu])
        precipitation.append(k.precipitation[enu])
        air_id.append(i)
        
    
    
```

```python
data = pd.DataFrame({'air_store_id':air_id,'precipitation':precipitation,'visit_date':calendar_date,'low_temperature':low_temperature,'avg_temperature':avg_temperature,'high_temperature':high_temperature})
```

```python
data.to_csv('weather_ready_to_use.csv',index = False)
```
