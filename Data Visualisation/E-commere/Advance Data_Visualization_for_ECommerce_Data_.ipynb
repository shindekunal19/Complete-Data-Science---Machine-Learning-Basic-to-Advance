```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file into a DataFrame
data = pd.read_csv("Copy of Power BI Dataset.csv")  # Replace 'your_file.csv' with your actual file path
```

```python
# Grouping by 'Country' and calculating total sales per country
sales_by_country = data.groupby('Country')['Sales'].sum().reset_index()

# Creating a bar plot to visualize total sales per country
plt.figure(figsize=(10, 6))
plt.bar(sales_by_country['Country'], sales_by_country['Sales'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.title('Total Sales by Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Importing Seaborn for enhanced visualizations

# Assuming 'data' contains the loaded CSV data (as shown in the previous example)

# Visualization 1: Bar plot - Total sales by Category
plt.figure(figsize=(8, 5))
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sns.barplot(x='Category', y='Sales', data=sales_by_category)
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization 2: Pie chart - Sales distribution by Segment
plt.figure(figsize=(7, 7))
sales_by_segment = data.groupby('Segment')['Sales'].sum().reset_index()
plt.pie(sales_by_segment['Sales'], labels=sales_by_segment['Segment'], autopct='%1.1f%%', startangle=140)
plt.title('Sales Distribution by Segment')
plt.tight_layout()
plt.show()

# Visualization 3: Line plot - Profit over time (Order Date)
data['Order Date'] = pd.to_datetime(data['Order Date'])
data.sort_values('Order Date', inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(data['Order Date'], data['Profit'], marker='o')
plt.title('Profit over Time')
plt.xlabel('Order Date')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Visualization 4: Scatter plot - Profit vs. Sales
plt.figure(figsize=(8, 6))
plt.scatter(data['Sales'], data['Profit'], color='orange', alpha=0.7)
plt.title('Profit vs Sales')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()

# Visualization 5: Box plot - Sales distribution by Region
plt.figure(figsize=(8, 6))
sns.boxplot(x='Region', y='Sales', data=data)
plt.title('Sales Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization 6: Histogram - Distribution of Quantity
plt.figure(figsize=(8, 6))
plt.hist(data['Quantity'], bins=10, edgecolor='black')
plt.title('Distribution of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Visualization 7: Heatmap - Correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()

# Visualization 8: Pairplot - Relationships between numerical columns
sns.pairplot(data[['Sales', 'Profit', 'Quantity', 'Cost']])
plt.suptitle('Pairplot of Sales, Profit, Quantity, and Cost', y=1.02)
plt.tight_layout()
plt.show()

# Visualization 9: Violin plot - Distribution of Profit by Category
plt.figure(figsize=(8, 6))
sns.violinplot(x='Category', y='Profit', data=data)
plt.title('Distribution of Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()

# Visualization 10: Stacked bar plot - Sales and Profit comparison by Sub-Category
sub_category_data = data.groupby('Sub-Category')[['Sales', 'Profit']].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(sub_category_data['Sub-Category'], sub_category_data['Sales'], label='Sales')
plt.bar(sub_category_data['Sub-Category'], sub_category_data['Profit'], label='Profit')
plt.xlabel('Sub-Category')
plt.ylabel('Amount')
plt.title('Sales and Profit by Sub-Category')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Visualization 11: Area plot - Sales over time (monthly)
data['Order Date'] = pd.to_datetime(data['Order Date'])
monthly_sales = data.resample('M', on='Order Date')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.fill_between(monthly_sales['Order Date'], monthly_sales['Sales'], color='skyblue', alpha=0.4)
plt.plot(monthly_sales['Order Date'], monthly_sales['Sales'], color='Slateblue', alpha=0.6)
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

# Visualization 12: Swarm plot - Profit distribution by Ship Mode
plt.figure(figsize=(8, 6))
sns.swarmplot(x='Ship Mode', y='Profit', data=data)
plt.title('Profit Distribution by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()

# Visualization 13: Pairwise relationships using pairplot (customized)
pairplot_custom = sns.pairplot(data[['Sales', 'Profit', 'Quantity', 'Cost']], kind='reg', diag_kind='kde')
pairplot_custom.fig.suptitle('Pairplot with Regression Lines and KDE', y=1.02)
plt.tight_layout()
plt.show()

# Visualization 14: Grouped bar plot - Sales and Profit comparison by Region and Segment
sales_profit_region_segment = data.groupby(['Region', 'Segment'])[['Sales', 'Profit']].sum().reset_index()
pivot_sales_profit = sales_profit_region_segment.pivot(index='Region', columns='Segment', values=['Sales', 'Profit'])
pivot_sales_profit.plot(kind='bar', figsize=(10, 6))
plt.title('Sales and Profit Comparison by Region and Segment')
plt.xlabel('Region')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend(title='Segment')
plt.tight_layout()
plt.show()

# Visualization 15: Radar chart - Comparison of Sales, Profit, and Cost by Sub-Category
sub_category_comparison = data.groupby('Sub-Category')[['Sales', 'Profit', 'Cost']].mean().reset_index()
categories = sub_category_comparison['Sub-Category']
values = sub_category_comparison.drop('Sub-Category', axis=1).values.transpose()

angles = [n / float(len(categories)) * 2 * 3.14159 for n in range(len(categories))]
angles += angles[:1]

plt.figure(figsize=(10, 8))
ax = plt.subplot(111, polar=True)
ax.set_theta_offset(0)
ax.set_theta_direction(-1)

plt.xticks(angles[:-1], categories, color='black', size=8)
ax.plot(angles, values[0], linewidth=1, linestyle='solid', label='Sales')
ax.fill(angles, values[0], 'skyblue', alpha=0.1)

ax.plot(angles, values[1], linewidth=1, linestyle='solid', label='Profit')
ax.fill(angles, values[1], 'salmon', alpha=0.1)

ax.plot(angles, values[2], linewidth=1, linestyle='solid', label='Cost')
ax.fill(angles, values[2], 'lightgreen', alpha=0.1)

plt.title('Comparison of Sales, Profit, and Cost by Sub-Category')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Visualization 16: Donut chart - Sales contribution by Category
plt.figure(figsize=(8, 8))
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
plt.pie(sales_by_category['Sales'], labels=sales_by_category['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Sales Contribution by Category')
plt.tight_layout()
plt.legend()
plt.show()

# Visualization 17: Bar plot - Sales comparison by City (top 10)
top_10_cities_sales = data.groupby('City')['Sales'].sum().nlargest(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Sales', data=top_10_cities_sales)
plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization 18: Word cloud - Product Name representation
from wordcloud import WordCloud

wordcloud_text = ' '.join(data['Product Name'].astype(str).tolist())
wordcloud = WordCloud(width=800, height=400, max_font_size=100, background_color='white').generate(wordcloud_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Product Names')
plt.tight_layout()
plt.show()

# Visualization 19: Violin plot - Profit distribution by Region and Category
plt.figure(figsize=(10, 6))
sns.violinplot(x='Region', y='Profit', hue='Category', data=data, split=True)
plt.title('Profit Distribution by Region and Category')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization 20: 3D scatter plot - Sales, Profit, and Quantity
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data['Sales'], data['Profit'], data['Quantity'], c='skyblue', marker='o')
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
ax.set_zlabel('Quantity')
ax.set_title('3D Scatter plot of Sales, Profit, and Quantity')
plt.tight_layout()
plt.show()
```

```python
# Visualization 21: Parallel Coordinates Plot - Comparison of numerical features
from pandas.plotting import parallel_coordinates

num_features = ['Sales', 'Profit', 'Quantity', 'Cost']
plt.figure(figsize=(10, 6))
parallel_coordinates(data[num_features + ['Category']], 'Category', colormap='viridis')
plt.title('Parallel Coordinates Plot of Numerical Features by Category')
plt.xlabel('Features')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()

# Visualization 22: Swarm plot - Sales distribution by Country and Segment
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Country', y='Sales', hue='Segment', data=data.sample(500))
plt.title('Sales Distribution by Country and Segment')
plt.xlabel('Country')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.legend(title='Segment')
plt.tight_layout()
plt.show()

# Visualization 23: Pairplot - Pairwise relationships colored by Segment
pairplot_segment = sns.pairplot(data[['Sales', 'Profit', 'Quantity', 'Cost', 'Segment']], hue='Segment')
pairplot_segment.fig.suptitle('Pairplot with Segment Color', y=1.02)
plt.tight_layout()
plt.show()

# Visualization 24: Distribution plot - Distribution of Profit by Region
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='Profit', hue='Region', multiple='stack')
plt.title('Distribution of Profit by Region')
plt.xlabel('Profit')
plt.ylabel('Count')
plt.legend(title='Region')
plt.tight_layout()
plt.show()

# Visualization 25: Bubble chart - Sales, Profit, and Quantity relationship by City
plt.figure(figsize=(10, 8))
plt.scatter(data['Sales'], data['Profit'], s=data['Quantity'] * 20, alpha=0.5)
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales, Profit, and Quantity Relationship by City')
plt.tight_layout()
plt.show()
```

```python
# Visualization 26: Donut chart - Category distribution
plt.figure(figsize=(8, 8))
category_counts = data['Category'].value_counts()
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Category Distribution')
plt.tight_layout()
plt.legend()
plt.show()

# Visualization 27: Box plot - Profit by Category and Segment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Profit', hue='Segment', data=data)
plt.title('Profit by Category and Segment')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.legend(title='Segment')
plt.tight_layout()
plt.show()

# Visualization 28: KDE plot - Distribution of Sales and Profit
plt.figure(figsize=(10, 6))
sns.kdeplot(data['Sales'], shade=True, label='Sales')
sns.kdeplot(data['Profit'], shade=True, label='Profit')
plt.title('Distribution of Sales and Profit')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()

# Visualization 29: Area plot - Cumulative Sales over time
monthly_sales_cumulative = data.resample('M', on='Order Date')['Sales'].sum().cumsum().reset_index()
plt.figure(figsize=(10, 6))
plt.fill_between(monthly_sales_cumulative['Order Date'], monthly_sales_cumulative['Sales'], color='orange', alpha=0.5)
plt.plot(monthly_sales_cumulative['Order Date'], monthly_sales_cumulative['Sales'], color='blue')
plt.title('Cumulative Sales over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Sales')
plt.tight_layout()
plt.show()

# Visualization 30: Pairwise comparison of numerical features with hexbin plot
sns.pairplot(data[['Sales', 'Profit', 'Quantity', 'Cost']], kind='hex')
plt.suptitle('Pairwise Comparison with Hexbin Plot', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Visualization 31: Stacked area plot - Profit over time by Region
region_profit_over_time = data.groupby(['Region', pd.Grouper(key='Order Date', freq='M')])['Profit'].sum().unstack('Region').fillna(0)
plt.figure(figsize=(10, 6))
region_profit_over_time.plot(kind='area', stacked=True)
plt.title('Profit over Time by Region')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()

# Visualization 32: Count plot - Ship Mode distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Ship Mode', data=data)
plt.title('Ship Mode Distribution')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Visualization 33: Scatter plot matrix - Relationships between numerical features
pd.plotting.scatter_matrix(data[['Sales', 'Profit', 'Quantity', 'Cost']], figsize=(10, 8))
plt.suptitle('Scatter Plot Matrix', y=1.02)
plt.tight_layout()
plt.show()

# Visualization 34: Treemap - Sales contribution by Sub-Category
import squarify

sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
plt.figure(figsize=(10, 8))
squarify.plot(sizes=sales_by_subcategory['Sales'], label=sales_by_subcategory['Sub-Category'], alpha=0.7)
plt.axis('off')
plt.title('Sales Contribution by Sub-Category (Treemap)')
plt.tight_layout()
plt.show()

# Visualization 35: 3D bar plot - Quantity comparison by Category and Region
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

category_region_quantity = data.groupby(['Category', 'Region'])['Quantity'].sum().reset_index()
for i, cat in enumerate(category_region_quantity['Category'].unique()):
    subset = category_region_quantity[category_region_quantity['Category'] == cat]
    xs = subset['Region']
    ys = [i] * len(subset)
    zs = subset['Quantity']
    ax.bar(xs, zs, zs=ys, zdir='y', label=cat)

ax.set_xlabel('Region')
ax.set_ylabel('Category')
ax.set_zlabel('Quantity')
ax.set_title('Quantity Comparison by Category and Region')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Visualization 36: Line plot - Sales trend by Category
category_sales_over_time = data.groupby(['Category', pd.Grouper(key='Order Date', freq='M')])['Sales'].sum().unstack('Category').fillna(0)
plt.figure(figsize=(10, 6))
category_sales_over_time.plot(kind='line')
plt.title('Sales Trend by Category over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()

# Visualization 37: Swarm plot - Sales distribution by Sub-Category and Segment
plt.figure(figsize=(12, 6))
sns.swarmplot(x='Sub-Category', y='Sales', hue='Segment', data=data.sample(500))
plt.title('Sales Distribution by Sub-Category and Segment')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.xticks(rotation=90)
plt.legend(title='Segment')
plt.tight_layout()
plt.show()

# Visualization 38: Box plot - Profit distribution by Region and Category
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Profit', hue='Category', data=data)
plt.title('Profit Distribution by Region and Category')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.legend(title='Category')
plt.tight_layout()
plt.show()

# Visualization 39: Jointplot - Sales vs. Profit with regression line
sns.jointplot(x='Sales', y='Profit', data=data, kind='reg', height=8)
plt.suptitle('Sales vs. Profit with Regression Line')
plt.tight_layout()
plt.show()

# Visualization 40: Pairplot - Pairwise relationships with hue based on Category
sns.pairplot(data, hue='Category')
plt.suptitle('Pairplot with Hue based on Category', y=1.02)
plt.tight_layout()
plt.show()

```

```python
# Visualization 41: Pie chart - Profit distribution by Segment
plt.figure(figsize=(8, 8))
profit_by_segment = data.groupby('Segment')['Profit'].sum()
plt.pie(profit_by_segment, labels=profit_by_segment.index, autopct='%1.1f%%', startangle=90)
plt.title('Profit Distribution by Segment')
plt.tight_layout()
plt.legend()
plt.show()

# Visualization 42: Box plot - Sales distribution by Region and Segment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='Sales', hue='Segment', data=data)
plt.title('Sales Distribution by Region and Segment')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(title='Segment')
plt.tight_layout()
plt.show()

# Visualization 43: Violin plot - Profit distribution by Category and Ship Mode
plt.figure(figsize=(10, 6))
sns.violinplot(x='Category', y='Profit', hue='Ship Mode', data=data)
plt.title('Profit Distribution by Category and Ship Mode')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.legend(title='Ship Mode')
plt.tight_layout()
plt.show()

# Visualization 44: Heatmap - Correlation matrix of numerical features
plt.figure(figsize=(8, 6))
sns.heatmap(data[['Sales', 'Profit', 'Quantity', 'Cost']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()

# Visualization 45: Pairplot - Pairwise relationships with different hues for Regions
sns.pairplot(data, hue='Region')
plt.suptitle('Pairplot with Hue based on Regions', y=1.02)
plt.tight_layout()
plt.show()
```
